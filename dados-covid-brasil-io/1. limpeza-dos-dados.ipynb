Agora que delimitamos o nosso dataset, precisamos verificar a consistência dos dados. Conforme explicado pelo Brasil IO, os registros dependem da divulgação dos municípios, o que pode não acontecer todos os dias. Para verificarmos se os registros são do dia, precisamos olhar algumas colunas:
* `is_repeated`:  Se o registro é referente ao boletim do dia ou foi copiado do último registro disponível.
* `is_last`: Se o registro é o mais novo para o local. 
* `last_available_date`: O último dia em que a cidade emitiu o boletim.


Vamos avaliar quais foram as ultimas atualizações por cidade.

```python
df_vale_do_paraiba.groupby('city')[['last_available_date']].last()
```

A última atualização dos dados na região foi 26/10/21. Isso quer dizer que todos os dados a frente dessa data mantém a mesma quantidade de casos. Como já se passaram mais de 2 meses e a recomendação no Brasil IO é utilizar dados de até 30 anterior à última atualização para se ter consistência dos dados, podemos disconsiderar os dados do mes de Outubro/21.

```python
#Transformando a coluna date em data
df_vale_do_paraiba['date'] = pd.to_datetime(df_vale_do_paraiba['date'], format='%Y-%m-%d')

#Filtrar o dataset para que tenha os registros até 30 de setembro de 2021
df_vale_2021 = df_vale_do_paraiba[df_vale_do_paraiba['date'] <= '2021-09-30']
df_vale_2021 = df_vale_2021.reset_index()
df_vale_2021.drop('index', axis=1, inplace=True)
df_vale_2021.head()
```

```python
df_vale_2021.info()
```

Agora vamos avaliar quais colunas serão úteis para a nossa análise, apoiado na [documentação do dataset](https://github.com/turicas/covid19-br/blob/master/api.md#caso_full)

```python
df_vale_2021[df_vale_2021['is_last'] == True]
```

```python
df_vale_2021[['place_type']].value_counts()
```

```python
df_vale_2021[['state']].value_counts()
```

As colunas a serem eliminadas:
* `city_ibge_code` traz o código IBGE do local - não será útil na nossa análse;
* `estimated_population_2019` traz a estimativa da população para a cidade/estado em 2019 segundo o IBGE, mas como a coluna possui dados desatualizados a própria documentação indica usar a coluna `estimated_population`;
* `is_last` indica se o registro é o mais novo para esse local, mas como quando extraímos a base de dados o registro mais novo era 26/10, para esse nosso sub dataset toda a coluna será falso, então podemos eliminar essa coluna.
* `place_type` indica se o registro se refere à uma cidade ou um estado. Como estamos lidando somente com estados, não vamos utilizar essa coluna;
* `state` traz a sigla da unidade federativa e como todas as cidades são do estado de São Paulo essa colna também não é necessária.

```python
colunas_para_retirar = ['city_ibge_code', 'estimated_population_2019', 'is_last', 
                        'place_type', 'state']
```

```python
df_vale_2021.drop(colunas_para_retirar, axis=1, inplace=True)
```

```python
#Salvar o dataset num novo arquivo csv.
df_vale_2021.to_csv('./dados/dados_limpos/vale_do_paraiba.csv', index=False, 
                    date_format='%Y-%m-%d', )
```
